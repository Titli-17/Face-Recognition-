**Haarcascade Classifier **

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Read Images

Img = cv2.imread("/content/sample_data/church-of-the-king-j9jZSqfH5YI-unsplash.jpg")
print(Img)
```

```python
print(Img.shape)
newImg = cv2.cvtColor(Img, cv2.COLOR_BGR2RGB)
```

```python
print(Img[0][0])
```

```python
# Show this Image
plt.imshow(Img)
plt.show()
```

```python
plt.imshow(newImg)
plt.axis("off")
plt.show()
```

```python
import cv2
model = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
faces = model.detectMultiScale(newImg,1.3,1)
faces
```

```python
for f in faces:
  x,y,w,h = f
  cv2.rectangle(newImg,(x,y),(x+w,y+h),(0,255,0),10)  #method for draw reactangle(10=thickness of the line)

plt.imshow(newImg)
plt.show()
```
